```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/magic/magic04.data', names=['LENGTH', 'WIDTH', 'SIZE', 'CONC', 'CONC1', 'ASYM', 'M3LONG', 'M3TRANS', 'ALPHA', 'DIST', 'CLASS'])
```

```python
data.info
```

```python
for feature in data.drop(['CLASS'], axis=1):
    sns.displot(data, x=data[feature], kind="kde", hue=data["CLASS"], palette="pastel")
```

```python
warnings.filterwarnings('ignore')

for feature in data.drop(["CLASS"], axis=1):
    for c in data["CLASS"].unique():
        sns.distplot(data[feature][data["CLASS"]==c],
                 kde=1,label='{}'.format(c))

    plt.legend()
    plt.show()
```
